# Lanczos

```python
n = 100
A = np.random.normal(size=(n,n))
A = A.T @ A
b = np.random.normal(size=n)
```

```python
k = 0
beta_prev = 0
r = b
q_prev = np.zeros(n)
x = np.zeros(n)
T = np.zeros((n,n))
maxits = n
for j in range(maxits):
    beta_curr = np.linalg.norm(r)
    q_curr = (1.0/beta_curr)*r
    k += 1
    v = A @ q_prev
    alpha = np.dot(q_prev, v)
    r = v - alpha*q_prev - beta_prev*q_prev
```
