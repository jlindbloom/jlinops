```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 10
n = 4
R = np.random.normal(size=(k,n))
```

```python
Rsvd = np.linalg.svd(R, full_matrices=True)
```

```python
U, S, Vt = Rsvd.U, np.vstack([np.diag(Rsvd.S), np.zeros( (k-n, n ) )]), Rsvd.Vh  
X = np.random.normal(size=(k,k))
```

```python
S.T @ U.T @ np.random.normal(size=(k,k)) @ U.T @ S
```

```python
k = 10
n = 5

R = np.random.normal(size=(k,n))
RtR = R.T @ R
Gt = np.linalg.cholesky(RtR)
G = Gt.T
GtG = G.T @ G
```

```python
Rsvd = np.linalg.svd(R, full_matrices=True)
Gsvd = np.linalg.svd(G, full_matrices=True)
```
