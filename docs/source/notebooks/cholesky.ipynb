```python
import numpy as np
import matplotlib.pyplot as plt

import jlinops 
```

```python
n = 10
A = np.random.normal(size=(n,n))
A = A.T @ A
Amat = A.copy()
A = jlinops.MatrixLinearOperator(A)
```

```python
Ainv = jlinops.BandedCholeskyInvOperator(A)
```

```python
Ainvmat = jlinops.black_box_to_dense(Ainv)
```

```python
np.linalg.norm(Ainvmat - np.linalg.inv(Amat))
```

```python
plt.imshow(np.linalg.inv(Ainv.chol_fac.toarray()))
plt.show()
```

# LinearOperator for the Cholesky factor?

```python
Linv = jlinops.BandedCholeskyFactorInvOperator(A)
```

```python
plt.imshow(jlinops.black_box_to_dense(Linv))
plt.show()
```

```python
plt.imshow(np.linalg.inv(Ainv.chol_fac.T.toarray()))
plt.show()
```

```python
plt.imshow(jlinops.black_box_to_dense(Linv.T))
plt.show()
```
