```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(t):
    if (t >= -2.8) and (t <= -2.1):
        return 2
    elif (t > -1.6) and (t <= -1.3):
        return 1.5
    elif t > 0:
        return 1.5*np.exp(  - ( (t - 0.5*np.pi)/(2/3) )**2)
    else:
        return 1
    
    
vfunc = np.vectorize(func)
```

```python
def func(tvec):
    output = np.ones_like(tvec)
    for i, t in enumerate(tvec):
        if (t >= -2.8) and (t <= -2.1):
            output[i] =  2
        elif (t > -1.6) and (t <= -1.3):
            output[i] = 1.5
        elif t > 0:
            output[i] = 1 + 1.5*np.exp(  - ( (t - 0.5*np.pi)/(2/3) )**2)
        else:
            output[i] = 1
    return output
```

```python
dom = np.linspace(-np.pi, np.pi, 200)
y = func(dom)
plt.plot(dom, y)
plt.show()
```
