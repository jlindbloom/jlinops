```python
from scipy.linalg import null_space
```

```python
# Size parameters
m = 20
k = 60
n = 50

np.random.seed(0)
# Draw random matrices
for j in range(100):

    # Build F and R
    F = np.random.normal(size=(m,n))
    R = np.random.normal(size=(k,n))

    # Make R have nontrivial kernel
    R[:,-1] = R[:,-3]
    R[:,2] = R[:,4] + 3*R[:,5] 

    # Build Q1
    Q1 = np.linalg.inv( (F.T @ F) + (R.T @ R) ) @ F.T

    # Build Q2
    Rpinv = np.linalg.pinv(R)
    W = null_space(R)
    FWpinv = np.linalg.pinv(F @ W)
    Roinv = (np.eye(n) - W @ FWpinv @ F) @ Rpinv
    
    second_part = W @ FWpinv
    first_part = Roinv @ np.linalg.inv( (Roinv.T @ F.T @ F @ Roinv) + np.eye(k) ) @ Roinv.T @ F.T

    Q2 = first_part + second_part

    print(np.linalg.norm( Q1 - Q2 ))
```
