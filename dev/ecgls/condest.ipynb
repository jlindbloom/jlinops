```python
import numpy as np
from scipy.sparse.linalg import eigsh, eigs
```

```python
m = 1000
n = 1000
F = np.random.normal(size=(m,n))
A = F.T @ F
```

```python
max_eigval, _ = eigsh(A, k=1, which="LM")
print(max_eigval)
```

```python
tau = 1.2*max_eigval
B = A - tau*np.eye(n)
smallest_eigval, _ = eigsh(B, k=1, which="LM")
```

```python
smallest_eigval + tau
```
