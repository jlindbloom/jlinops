```python
import numpy as np
import matplotlib.pyplot as plt

import jlinops
```

```python
m, n = 5, 7

Rd = jlinops.Dirichlet2DSym((m,n))
Rn = jlinops.Neumann2D((m,n))

Rd = jlinops.black_box_to_dense(Rd)
Rn = jlinops.black_box_to_dense(Rn)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(8,5))

axs[0].imshow(Rd.T @ Rd)
axs[0].set_title("Dirichlet BC")

axs[1].imshow(Rn.T @ Rn)
axs[1].set_title("Neumann BC")

plt.show()
```
